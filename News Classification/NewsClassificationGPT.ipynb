### Get the full headline string 
* Since, only getting headlines index from GPT (reducing token)
* This code fetchs index from the GPT output and then uses `data_list` to get the headline string

```python

# Initialize an empty list to store the final results
final = []

# Initialize an empty list to store error entries
error_list = []

# Loop through each item in result_med
for item in result_med:
    # Split the item into lines
    lines = item.split(';')
    
    # Extract the headline index from the first line if it's available, otherwise skip the entry
    try:
        index = int(lines[0].split(':')[0].strip())
    except ValueError:
        error_list.append(item)
        continue
    
    # Use the index to fetch the actual headline from data_list if it exists, otherwise skip the entry
    if 0 <= index < len(data_list):
        actual_headline = data_list[index]['news']['head']
    else:
        error_list.append(item)
        continue
    
    # Create a new list containing the actual headline and the rest of the contents from result_med
    result = [actual_headline] + [line.strip() for line in lines]
    
    # Append the result to the final list
    final.append(result)

# Print the final list
#for item in final:
    #print(item)

# Print the error entries
#for error_item in error_list:
    #print("Error Entry:", error_item)

final[:10]
```
