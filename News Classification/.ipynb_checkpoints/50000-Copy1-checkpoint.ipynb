```python
def find_missing_elements(final2, test2):
    missing_elements = set()
    
    for element in test2:
        found = False
        for key, values in final2.items():
            if element in values:
                found = True
                break
        if not found:
            missing_elements.add(element)
    
    return missing_elements
```
